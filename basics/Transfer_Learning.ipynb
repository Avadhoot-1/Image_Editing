```python
import torch
import torchvision
import torch.nn.functional as F
from torchvision import datasets, transforms

transform = transforms.Compose([ transforms.CenterCrop(224),
                                 transforms.ToTensor()])

train_data = datasets.ImageFolder('train/', transform=transform)
val_data = datasets.ImageFolder('Val/', transform=transform)
train_loader = torch.utils.data.DataLoader(train_data, batch_size=32, shuffle=True)
test_loader = torch.utils.data.DataLoader(val_data, batch_size=32, shuffle=True)
```

```python


import torch
import torch.nn as nn
import torchvision.models as models
```

```python
class New_model(nn.Module):

  def __init__(self, pre_model, classes):
    super().__init__()

    self.pre = pre_model
    self.pre.classifier = nn.Sequential(
        nn.Linear(512*7*7, 128),
        nn.ReLU(),
        nn.Dropout(p = 0.2),
        nn.Linear(128, classes)
    )

  def forward(self, x):
    x = self.pre(x)
    return x
```

```python
from torchsummary import summary
import torch.nn as nn
import torchvision.models as models
from tqdm import tqdm


if torch.cuda.is_available():
    device = torch.device("cuda")
    print("CUDA is available! Using GPU for computations.")
else:
    device = torch.device("cpu")
    print("CUDA is not available. Using CPU for computations.")

pre_model = None
pre_model = torch.hub.load('pytorch/vision:v0.10.0', 'vgg11', pretrained=True)
# pre_model = pre_model.cuda()
for param in pre_model.parameters():
  param.requires_grad = False
for title, param in pre_model.named_parameters():
  if title.startswith('features.18'):
    param.requires_grad = True




model = New_model(pre_model, 200)
model.to(device)
summary(model, (3,224,224))
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(),lr= 3e-3)
```

```python

loss_list = []
val_loss = []
val_acc = []
for epoch in range(10):
    training_loss = 0
    val_loss_item = 0
    model.train()
    for inputs, labels in tqdm(train_loader, desc="Training"):
        inputs = inputs.to(device)
        labels = labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        training_loss += loss.item()
        loss.backward()
        optimizer.step()

    if(epoch%1 == 0):
        correct = 0
        total = 0
        for inputs, labels in tqdm(test_loader, desc="validation"):
            inputs = inputs.to(device)
            labels = labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            max_val, predicted = outputs.max(1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
            loss = criterion(outputs, labels)
            val_loss_item += loss.item()
        val_loss.append(val_loss_item/ len(test_loader))
        val_acc.append(correct/total)
        print(f'Validation epoch {epoch} acuraccy : {correct/total}, loss: {val_loss_item/len(test_loader)}')
    loss_list.append(training_loss / len(train_loader))
    print(f'Epoch [{epoch + 1}/15], Loss: {training_loss / len(train_loader):.4f}')
```

```python
plt.style.use("seaborn")

plt.plot(loss_list,label="training loss",c='red')
plt.plot(val_loss,label="validation loss",c='blue')
plt.legend()
plt.show()
```
